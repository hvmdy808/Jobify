## Initial test
For testing purposes, I will use 'data' as a query in the URL to see more results than using a proper job title.
Now, I will use a session with a header to simulate a request coming from a real browser in order to not get blocked.
If the status code is 200, i will get the number of jobs returned. The title is written like '3,404 data jobs in Egypt â€“ discover job details now on Wuzzuf!'
All job cards that have the content we need are divs with class 'css-ghe2tq', and we will use them to parse any needed info from them.

```python
url = "https://wuzzuf.net/search/jobs/?q=data"
session = requests.Session()

session.headers.update({
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://www.google.com/",
    "Accept-Encoding": "gzip, deflate, br",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Connection": "keep-alive"
})

try:
    time.sleep(2)
    response = session.get(url, timeout=15)
    response.raise_for_status()
except requests.exceptions.HTTPError as http_err:
    if response.status_code == 403:
        print("Access forbidden. The website is blocking automated requests.")
    else:
        print(f"HTTP error occurred: {http_err}")
else:
    print("Request successful!")
    print(response.status_code)
    print(response.content)
    soup = BeautifulSoup(response.content, 'html.parser')
    title = soup.find('title').contents[0]
    print(title)
    num = ''
    title_text = title.text
    for i in range(len(title_text)):
        if title_text[i] == ' ':
            break
        if title_text[i] == ',':
            pass
        else:
            num += title_text[i]

    print(int(num) - 1)


    cards = soup.find_all('div', {'class':'css-ghe2tq'})
    print(cards)
    print(len(cards))
```

## Parsing titles

```python
titles: list[str] = []
for i in range(len(cards)):
    title = cards[i].find('a', {'rel':'noreferrer', 'class':'css-o171kl'}).get_text('')
    titles.append(title.strip())

print(titles)
print(len(titles))
```

## Parsing companies

```python
companies: list[str] = []
for i in range(len(cards)):
    company = cards[i].find('a', {'class':'css-ipsyv7'}).get_text('')
    companies.append(company.strip())

print(companies)
print(len(companies))
```

## Parsing locations
Used get_text with argument '' to get all the text in the span even if it is striped, the '' will concatenate them with empty string as a delimiter

```python
locs: list[str] = []
for i in range(len(cards)):
    loc = cards[i].find('span', {'class':'css-16x61xq'}).get_text('')
    locs.append(loc.strip())

print(locs)
print(len(locs))
```

### Final working one
Avoid using .contents if possible

```python
exp_levels: list[str] = []
for i in range(len(cards)):
    parent = cards[i].find('div', {'class':'css-1rhj4yg'})
    divs = parent.find_all("div", recursive=False)
    div = divs[1]
    a = div.find('a')
    exp = a.get_text('')
    exp_levels.append(exp.strip())

print(exp_levels)
print(len(exp_levels))
```

## Parsing years of experience

```python
years: list[str] = []
for i in range(len(cards)):
    parent = cards[i].find('div', {'class':'css-1rhj4yg'})
    divs = parent.find_all("div", recursive=False)
    div = divs[1]
    span = div.find('span')
    year = span.get_text('')
    years.append(year.strip())

print(years)
print(len(years))
```

## Parsing skills & keywords for each job

```python
skills = []
for i in range(len(cards)):
    parent = cards[i].find('div', {'class':'css-1rhj4yg'})
    divs = parent.find_all("div", recursive=False)
    div = divs[1]
    a_list = div.find_all('a', recursive = False)
    a_list = a_list[1:]
    skill_coll: list[str] = []
    for j in range(len(a_list)):
        skill = a_list[j].get_text(strip = True)
        skill_coll.append(skill)

    skills.append(skill_coll)

for i in range(len(skills)):
    print(skills[i])
print(len(skills))
```
